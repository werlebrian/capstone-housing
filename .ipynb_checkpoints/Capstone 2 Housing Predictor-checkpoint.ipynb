```python
# Capstone Project: Housing Data EDA and Wrangling

This notebook performs exploratory data analysis (EDA) and initial data wrangling on the King County housing dataset.

**Objectives:**
- Load and inspect the dataset
- Perform initial cleaning
- Explore numerical and categorical features
- Check for missing values and duplicates
- Investigate outliers
- Prepare dataset for future modeling
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
file_path = r'C:\Users\werle\Desktop\Python_Projects\capstone-housing\data\raw\kc_house_data.csv'
df = pd.read_csv(file_path)

# Checking the data
print(df.sample(5))
print(df.head())
print(df.shape)
```

```python
df.info()
```

```python
# Drop 'id' and 'date', not needed for analysis
df.drop(['id', 'date'], axis='columns', inplace=True)

# Check for remaining category/object columns
object_cols = df.select_dtypes('object').columns.tolist()
print(object_cols)
```

```python
print(df.describe())

# Categorical summary (if any)
if object_cols:
    for col in object_cols:
        print(f"\nCategorical Summary: {col}")
        print(df[col].value_counts())
        print(df[col].value_counts(normalize=True))
```

```python
# Missing values, if any
print(df.isnull().sum())
```

```python
duplicate_count = df.duplicated().sum()
print(duplicate_count)

# Remove duplicates if any
if duplicate_count > 0:
    df.drop_duplicates(inplace=True)
    print(df.shape)
```

```python
#Outlier exploration

numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

for col in numerical_cols:
    print(col)

    # Boxplot
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()

    # Outlier detection using IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    print(len(outliers))
```

```python
def plot_corr(df, figsize=(16, 12)):
    corr = df.corr()
    
    _, ax = plt.subplots(1, 1, figsize=figsize)
    g = sns.heatmap(corr, ax=ax, annot=True, cmap=sns.cubehelix_palette(start=.5, rot=-.5, as_cmap=True))

    for _ax in g.get_xticklabels():
        _ax.set_rotation(75)
    
    
plot_corr(df)
```

```python
def plot_base_relation(df, figsize=(20, 200)):
    columns = df.columns.tolist()
    _, axs = plt.subplots(len(columns), 4, figsize=figsize)
    
    for idx, column in enumerate(columns):
        # Check Distribution
        sns.histplot(
            x=df[column],
            kde=False,
            color='#65b87b', alpha=.7,
            ax=axs[idx][0]
        )

        # Checking outliers
        sns.boxplot(
            x=df[column],
            color='#6fb9bd',
            ax=axs[idx][1]
        )

        # Relationship to price
        sns.scatterplot(
            x=column, y='price', data=df,
            color='#706dbd', alpha=.7, s=80,
            ax=axs[idx][2]
        )
        
        # count 'column'
        sns.countplot(
            x=column, data=df,
            color='#42b0f5', alpha=.7,
            ax=axs[idx][3]
        )
        
        
plot_base_relation(df, figsize=(20, 70))
```
